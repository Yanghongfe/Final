```python

import pandas as pd
```

```python
df= pd.read_csv('features_data.csv')
df.groupby('race').agg({'name': 'count'})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# 设置Seaborn主题为白色背景
sns.set_theme(style='white')
# 创建FacetGrid对象，指定col参数将种族分开到不同的列
g = sns.FacetGrid(df, col="race", col_wrap=4, height=3, aspect=1)
# 使用map方法添加直方图
g.map(sns.histplot, "length", kde=False, binwidth=1, alpha=0.6)
# 添加图例
g.add_legend(title="Race")
# 设置图表标题和轴标签
g.fig.suptitle("Distribution Plot of Length by Race", fontsize=18)
# 调整标题位置
plt.subplots_adjust(top=0.85)
# 显示图形
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
g = sns.FacetGrid(df, col="race", col_wrap=4, height=4, aspect=1)
g.map(sns.histplot, "fl_is_v", kde=False, binwidth=0.5, alpha=0.8)
g.fig.suptitle("Distribution Plot of fl_is_v by Race", fontsize=18, y=1.05)
g.set_axis_labels("FL_is_V Value", "Count")
for ax, title in zip(g.axes.flat, df['race'].unique()):
    ax.set_title(title)
plt.tight_layout()


plt.show()

```
